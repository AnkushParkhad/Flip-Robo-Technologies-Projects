# 1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos. You need to find following details: A) Rank B) Name C) Artist D) Upload date E) Views

```python
#importing required libraries
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import warnings
warnings.filterwarnings('ignore')

#Now lets first connect to the web driver
driver=webdriver.Chrome(r'C:\Users\ankush.parkhad\Desktop\data science\chrome driver\chromedriver.exe')

#opening the given link
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')

driver.maximize_window()

#creating empty lists to store our scraped data
rank=[]
name=[]
artist=[]
upload_date=[]
views=[]

#scrapping rank of all videos
ranks=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
for i in ranks:
    rank.append(i.text)

#scrapping name of all videos
names=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
for i in names:
    name.append(i.text)

#scrapping artists of all videos
artists=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
for i in artists:
    artist.append(i.text)

#scrapping uploading date
date=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
for i in date:
    upload_date.append(i.text)

#scrapping views(in billions)
view=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
for i in view:
    views.append(i.text)

#creating a dataframe to store the values
df=pd.DataFrame({'Rank':rank, "Name": name, 'Artist':artist, 'Publication date':upload_date, 'Views (in billions)': views})
df
```

 # 2. Scrape the details team India’s international fixtures from bcci.tv. Url = https://www.bcci.tv/. You need to find following details: A) Match title (I.e. 1st ODI) B) Series C) Place D) Date E) Time

```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import time

#connect to the web driver
driver=webdriver.Chrome(r'C:\Users\ankush.parkhad\Desktop\data science\chrome driver\chromedriver.exe')


#opening the given link
driver.get('https://www.bcci.tv/')

driver.maximize_window()

#clicking on menu button
menu=driver.find_element(By.XPATH,'/html/body/nav/div[1]/button')
menu.click()

#clicking on international button
element=driver.find_element(By.XPATH,'/html/body/nav/div[1]/div[2]/ul[1]/li[2]/a')
element.click()

#clicking on more fixtures button
button=driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div/div/div/div[2]/div[3]/div[2]/div/button')
button.click()

#clicking once again on more fixture button to load all matches
button=driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div/div/div/div[2]/div[3]/div[2]/div/button')

#creating empty lists to store all data
title=[]
series=[]
place=[]
date=[]
time=[]

#scrapping titles
titles=driver.find_elements(By.XPATH,'//span[@class="matchOrderText ng-binding ng-scope"]')
for i in titles:
    title.append(i.text.split("-")[0][:-1])

#scrapping series
seriess=driver.find_elements(By.XPATH,'//span[@class="ng-binding"]')
for i in seriess:
    series.append(i.text)

#scrapping place
location=driver.find_elements(By.XPATH,'//div[@class="fix-place ng-binding ng-scope"]')
for i in location:
    place.append(i.text.split('-')[1])

#scrapping date
datee=driver.find_elements(By.XPATH,'//h5[@class="ng-binding"]')
for i in datee:
    date.append(i.text)

#scrapping time
timee=driver.find_elements(By.XPATH,'//div[@class="match-card-right match-schedule "]')
for i in timee:
    time.append(i.text)

#creating dataframe
df=pd.DataFrame({'Match Title':title, 'Series':series, 'Place':place, 'Date':date, 'Time':time})

df
```

# 9. Scrape the details most watched tv series of all time from imdb.com.
Url = https://www.imdb.com/list/ls095964455/
You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes

```python
#importing required libraries
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import warnings
warnings.filterwarnings('ignore')

#Now lets first connect to the web driver
driver=webdriver.Chrome(r'C:\Users\ankush.parkhad\Desktop\data science\chrome driver\chromedriver.exe')

#opening the given link
driver.get('https://www.imdb.com/list/ls095964455/')
driver.maximize_window()

#creating empty lists to store data
name=[]
year=[]
gen=[]
time=[]
ratings=[]
votes=[]

#scrapping names
names=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/h3/a')
for i in names:
    name.append(i.text)

#scrapping years
years=driver.find_elements(By.XPATH,'//span[@class="lister-item-year text-muted unbold"]')
for i in years:
    year.append(i.text)

#scrapping genre
genre=driver.find_elements(By.XPATH,'//span[@class="genre"]')
for i in genre:
    gen.append(i.text)

#scrapping runtime
run=driver.find_elements(By.XPATH,'//span[@class="runtime"]')
for i in run:
    time.append(i.text)

#scrapping ratings
rate=driver.find_elements(By.XPATH,'//div[@class="ipl-rating-star small"]')
for i in rate:
    ratings.append(i.text)

#scrapping votes
vote=driver.find_elements(By.XPATH,'//span[@name="nv"]')
for i in vote:
    votes.append(i.text)

#creating dataframe
df=pd.DataFrame({'Name':name, 'Year Span':year, 'Genre':gen, 'Run Time':time, 'Ratings':ratings, 'Votes':votes})
df
```
